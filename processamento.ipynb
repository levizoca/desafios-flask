# Python

```python
### Importando bibliotecas

import pandas as pd
```

```python
### Lendo arquivo e renomeando colunas

filename = 'K3241.K03200Y9.D30211.ESTABELE.csv'
columns = ['CNPJ BÁSICO',
           'CNPJ ORDEM ',
           'CNPJ DV',
           'IDENTIFICADOR MATRIZ/FILIAL',
           'NOME FANTASIA',
           'SITUAÇÃO CADASTRAL',
           'DATA SITUAÇÃO CADASTRAL',
           'MOTIVO SITUAÇÃO CADASTRAL',
           'NOME DA CIDADE NO EXTERIOR',
           'PAIS',
           'DATA DE INÍCIO ATIVIDADE ',
           'CNAE FISCAL PRINCIPAL',
           'CNAE FISCAL SECUNDÁRIA',
           'TIPO DE LOGRADOURO',
           'LOGRADOURO',
           'NÚMERO',
           'COMPLEMENTO',
           'BAIRRO',
           'CEP ',
           'UF',
           'MUNICÍPIO',
           'DDD 1 ',
           'TELEFONE 1',
           'DDD 2',
           'TELEFONE 2',
           'DDD DO FAX',
           'FAX ',
           'CORREIO ELETRÔNICO',
           'SITUAÇÃO ESPECIAL',
           'DATA DA SITUAÇÃO ESPECIAL']
df = pd.read_csv(filename, sep=';', names = columns, encoding='latin-1')
```

```python
### Vendo numero de linhas e colunas

df.shape
```

```python
### Criando colunas CNPJ COMPLETO e TELEFONE

df['CNPJ COMPLETO'] = df['CNPJ BÁSICO'].astype(str) + df['CNPJ ORDEM '].astype(str) + df['CNPJ DV'].astype(str)
df['TELEFONE'] = df['DDD 1 '].astype(str) + df['TELEFONE 1'].astype(str)
```

```python
### separado colunas

dfreg = df.sample(10)
dfreg.sample(10)
```

```python
### Criando pivot_table apenas com as colunas que serão utilizadas

registros = dfreg.pivot_table(
    index='CNPJ COMPLETO',
    values=['NOME FANTASIA', 'CEP ', 'TELEFONE', 'CORREIO ELETRÔNICO'],
    aggfunc='first'
)
registros.sample(10)
```

```python
### Salvando arquivo em csv com apenas 10 registros

registros.sample(10).to_csv('estabelecimentos10.csv')
```

```python
### Transformando arquivo csv em json

df = pd.read_csv('estabelecimentos10.csv')
df = df.fillna('None')
estabelecimentos_data = df.to_dict(orient = "records")
estabelecimentos_data
```

# Elastic Search

```python
### Importando bibliotecas para conexão com o Elasticsearch

import elasticsearch
print(elasticsearch.__version__)
```

```python
### Conectando ao Elasticsearch

es = elasticsearch.Elasticsearch([{'host': 'localhost', 'port': 9200}])
resp = es.info()
```

```python
### Inserindo dados no Elasticsearch

estabelecimentos_data = df.to_dict(orient = "records")
for data in estabelecimentos_data: 
  es.index(index='estabelecimentos', document=data)
```

```python
### Buscando dados no Elasticsearch

res = es.search(index="estabelecimentos", body={"query": {"match_all": {}}})
print("Got %d Hits:" % res['hits']['total']['value'])

for hit in res['hits']['hits']:
    print("%(CNPJ COMPLETO)s %(NOME FANTASIA)s: %(TELEFONE)s" % hit["_source"])
```
